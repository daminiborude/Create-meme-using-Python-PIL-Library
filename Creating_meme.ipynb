## Loading Image using Matplotlib

```python
import matplotlib.pyplot as plt
from matplotlib import image
```

```python
import PIL 
from PIL import Image, ImageDraw, ImageFont
```

```python
img = Image.open('meme_image.jpg') 
print(type(img), img.size)
plt.imshow(img)
```

```python
print("Image File Name:", img.filename)
print("Shape/Size of Image:", img.size) # (Width, Height in pixels)
print("Image Mode:", img.mode)
print("Image Format:", img.format)
```

```python
Draw = ImageDraw.Draw(img)
Text = "WEEKEND"

font = ImageFont.truetype('static/Inconsolata_Condensed-Bold.ttf', 100)

Draw.text((480, 180), text=Text, fill=(0, 0, 0), font=font, align="center")

plt.imshow(img)
```
